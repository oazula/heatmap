### UFO heat map

```python
import pandas as pd 
import folium
from folium.plugins import HeatMap 
```

```python
df=pd.read_csv('ufo_sighting_data.csv')
df.head()
```

```python
usa_data=df[df['country']=='us']#para usar solo los datos de us
#usa_data.head()
```

```python
virginia=usa_data[usa_data['state/province']=='wv']
#virginia.head()
```

Creamos 2 mapas

```python
map=folium.Map(location=[38.5, -80.5], zoom_start=8)#los marcadores 
heat_m=folium.Map(location=[38.5, -80.5], zoom_start=8)
# Create a new Folium map centered!! at latitude 38.5, longitude 
# -80.5, with an initial zoom level of 8
```

***Quitando los NAN:**

```python
virginia=virginia.dropna()
```

Hasta ahora tenemos el mapa sin los datos e hicimos una peque√±a limpieza sobre los dato
ahora A agregarlos 

```python
for index, row in virginia.iterrows():#vamos a agregar un marcador!!
    folium.CircleMarker(location=[row['latitude'], row['longitude']], radius=10, 
                        popup=row['description'], fill_color= 'green').add_to(map)#El marcador va a tener esas coordenadas
```

Miremos el heatmap

```python
virginia_mat=virginia[['latitude', 'longitude']].to_numpy()
#lo converti en matriz 2D cuyos valores son parehjas de latitud longitud

HeatMap(virginia_mat).add_to(heat_m)
#creo el heatmap al mapa que no tiene los marcadores 
```
